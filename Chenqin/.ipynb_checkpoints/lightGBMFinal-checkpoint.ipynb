```python
init_record['_key_occurreddate_month']=init_record['_key_occurreddate_month'].astype("datetime64[ns]")
init_record = init_record.set_index(['_key_client_id', '_key_occurreddate_month'])
```

## Split Dataset to  Training and evaluation

```python
import random
clientList = list(set(init_record.index.get_level_values(0)))
```

```python
random.shuffle(clientList)
```

```python
length = len(clientList)
ratio = int(length*0.2)
trainIDs = clientList[ratio:]
testIDs = clientList[:ratio]
train = init_record.drop(testIDs, level=0)
evl = init_record.drop(trainIDs, level=0)
```

```python
print(train.shape)
print(evl.shape)
```

```python
import numpy as np
```

```python

def construct_feature_label(data):
    y = []
    clientList = list(set(data.index.get_level_values(0)))
    data['response_variable'] = data.groupby(level=0)['response_variable'].shift(-1)  
    data = data.groupby(level=0, as_index=False).apply(lambda group: group.iloc[:-1])
    #print(data.loc["0000D172-EA88-432F-8235-9FAA00D29072"]['response_variable'])
    y = data['response_variable'].astype('int32')
    x = data.drop('response_variable',axis='columns')
    return x, y

x_train, y_train = construct_feature_label(train)
x_eval, y_eval = construct_feature_label(evl)
```

```python
print(len(y_train))
print(x_train.shape)
print(len(y_eval))
print(x_eval.shape)
```

# Start training and predictions

```python
from sklearn.metrics import f1_score, confusion_matrix, roc_curve, accuracy_score
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import datetime
```

```python
train_data = lightgbm.Dataset(x_train, label=y_train)
eval_data = lightgbm.Dataset(x_eval, label=y_eval)
```

```python
mdl = lightgbm.LGBMClassifier(boosting_type= 'gbdt', 
          objective = 'binary', 
          n_jobs = 5, 
          silent = True)

# To view the default model parameters:
mdl.get_params().keys()
```

```python
grid = GridSearchCV(mdl, gridParams, verbose=1, cv=5, n_jobs=-1)
```

```python
grid.fit(x_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
parameters = {
    'application': 'binary',
    'objective': 'binary',
    'metric': 'auc',
    'is_unbalance': 'true',
    'boosting': 'gbdt',
    'num_leaves': 31,
    'feature_fraction': 0.8,
    'bagging_freq': 20,
    'learning_rate': 0.05,
}
```

```python
a = datetime.datetime.now()
model = lightgbm.train(parameters,
                       train_data,
                       valid_sets=eval_data,
                       num_boost_round=2000)
b = datetime.datetime.now()
```

## Evaluation

```python
from sklearn.metrics import classification_report

report = classification_report(y_eval, y_hat, labels=[0,1])
print(report)
```

### Model ensembling

```python
evl_scores1 = model2.predict(x_eval)
evl_scores2 = model3.predict(x_eval)
evl_scores3 = model.predict(x_eval)


```

```python
conc = []
conc.append(evl_scores1)
conc.append(evl_scores2)
conc.append(evl_scores3)
conc = np.array(conc)
print(len(evl_scores1))
print(len(evl_scores2))
print(len(evl_scores3))
```

```python
conc
```

```python
# mean vote
y_hat_ensemble_model = np.where(conc.mean(axis=0) < 0.5, 0, 1) 
f1_score(y_eval,y_hat_ensemble_model)
```

```python
# majority vote
y_hat = []
for i in range(len(conc[0])):
    count=0
    for j in range(len(conc)):
        if conc[j][i] < 0.5:
            count+=1
    if count<=len(conc)//2:
        y_hat.append(1)
    else:
        y_hat.append(0)
            
f1_score(y_eval,y_hat)  
```

```python
## test data

test_scores1 = model2.predict(x_test)
test_scores2 = model3.predict(x_test)
test_scores3 = model.predict(x_test)
```

```python
conc = []
conc.append(test_scores1)
conc.append(test_scores2)
conc.append(test_scores3)
conc = np.array(conc)

y_hat_ensemble_model = np.where(conc.mean(axis=0) < 0.5, 0, 1) 
f1_score(y_test,y_hat_ensemble_model)
```
