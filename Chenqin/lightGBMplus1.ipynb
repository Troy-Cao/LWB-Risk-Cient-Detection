```python
import lightgbm
import pandas as pd

part_1 = pd.read_csv("../../Data/dc_part_v3.csv")
```

```python
part_2 = pd.read_csv("../../Data/chenqin5%.csv")
part_2.shape
part_2
```

```python
part_3 = pd.read_csv("../../Data/miaoqin_partN.csv")
part_3 = part_3.drop(["Unnamed: 0"], axis='columns')
print(part_3.shape)
part_3
```

```python
part_4 = pd.read_csv("../../Data/yunDataN.csv")
part_4 = part_4.drop(["Unnamed: 0"], axis='columns')
part_4
```

## Data preprocessing

```python
init_record = pd.concat([part_1,part_2,part_3, part_4], axis=1, sort=False)
init_record.head()
print(init_record.shape)
```

#### remove unwanted columns

```python
sum_n_days = init_record.filter(regex=r'_sum_n_days$', axis=1).columns
avg_n_days = init_record.filter(regex=r'_avg_n_days$', axis=1).columns
days_since = init_record.filter(regex=r'_days_since$', axis=1).columns
lag = init_record.filter(regex=r'_lag[0-9]*$', axis=1).columns

init_record = init_record.drop(sum_n_days,axis=1)
init_record = init_record.drop(avg_n_days,axis=1)
init_record = init_record.drop(days_since,axis=1)
init_record = init_record.drop(lag,axis=1)
```

```python

#init_record = pd.concat([part_1, part_2], axis=1, sort=False)
init_record['_key_occurreddate_month']=init_record['_key_occurreddate_month'].astype("datetime64[ns]")
init_record = init_record.set_index(['_key_client_id', '_key_occurreddate_month'])
```

## Split Dataset to three part: Training, evaluation and testing data

```python
from collections import Counter
count = Counter()
for i in init_record.index.get_level_values(0):
    count[i] += 1
count.most_common(1)
```

```python
import random
clientList = list(set(init_record.index.get_level_values(0)))
```

```python
random.shuffle(clientList)
```

```python
length = len(clientList)
ratio = int(length*0.1)
trainIDs = clientList[ratio:]
testIDs = clientList[:ratio]
train0 = init_record.drop(testIDs, level=0)
test = init_record.drop(trainIDs, level=0)
```

```python
print(train0.shape)
print(test.shape)
```

```python
ratio = int(len(trainIDs)*0.25)
trainRatio = trainIDs[ratio:]
evalRatio  = trainIDs[:ratio]
train = train0.drop(evalRatio, level=0)
evl = train0.drop(trainRatio,level=0)
```

```python
print(train.shape)
print(evl.shape)
print(test.shape)
```

## Construct feature and label

```python
import numpy as np
```

#### <span style="color:red">Notice! This function will change dataframe inplace, if running second time have to rerun the code of init_record  </span>

```python

def construct_feature_label(data):
    y = []
    clientList = list(set(data.index.get_level_values(0)))
    data['response_variable'] = data.groupby(level=0)['response_variable'].shift(-1)  
    data = data.groupby(level=0, as_index=False).apply(lambda group: group.iloc[:-1])
    #print(data.loc["0000D172-EA88-432F-8235-9FAA00D29072"]['response_variable'])
    y = data['response_variable']
    x = data.drop('response_variable',axis='columns')
    return x, y

x_train, y_train = construct_feature_label(train)
x_eval, y_eval = construct_feature_label(evl)
```

```python
print(len(y_train))
print(x_train.shape)
print(len(y_eval))
print(x_eval.shape)
```

# Start training and predictions

```python
from sklearn.metrics import f1_score
from sklearn import metrics
```

```python
train_data = lightgbm.Dataset(x_train, label=y_train)
eval_data = lightgbm.Dataset(x_eval, label=y_eval)
```

```python
parameters = {
    'application': 'binary',
    'objective': 'binary',
    'metric': 'auc',
    'is_unbalance': 'true',
    'boosting': 'gbdt',
    'num_leaves': 31,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 20,
    'learning_rate': 0.05,
    'verbose': 0
}
```

```python
model = lightgbm.train(parameters,
                       train_data,
                       valid_sets=eval_data,
                       num_boost_round=2000)
```

```python
y_hat = np.where(model.predict(x_eval) < 0.5, 0, 1)  
f1_score(y_eval,y_hat)
```

# Cross validation

```python
model_cv = lightgbm.cv(parameters,
                       train_data,
                       num_boost_round=2000,return_cvbooster=True)
```

```python
cv_result = model_cv["cvbooster"].predict(x_eval)
```

```python
cv_result = np.array(cv_result)
```

#### Mean vote

```python
y_hat = np.where(cv_result.mean(axis=0) < 0.5, 0, 1) 
f1_score(y_eval,y_hat)
```

#### Majority vote

```python
y_hat = []
for i in range(len(cv_result[0])):
    count=0
    for j in range(len(cv_result)):
        if cv_result[j][i] < 0.5:
            count+=1
    if count<=len(cv_result)//2:
        y_hat.append(1)
    else:
        y_hat.append(0)
            
f1_score(y_eval,y_hat)  
```

#### One vote

```python
y_hat = []
for i in range(len(cv_result[0])):
    count=0
    for j in range(len(cv_result)):
        if cv_result[j][i] > 0.5:
            count+=1
    if count>0:
        y_hat.append(1)
    else:
        y_hat.append(0)
            
f1_score(y_eval,y_hat)  
```
