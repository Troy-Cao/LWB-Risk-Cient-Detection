```python
fullData = pd.read_csv(filename)
```

### Get Categories

Divide categoricals by the first starting word

```python
dataCateg = []
for i in fullData.columns:
    categ = i.split("_")[0] + "_"
    if categ not in dataCateg:
        dataCateg.append(categ)
```

```python
print(dataCateg)
```

For example, search category start with 'occurreddate_' in data, we could simply use 
`fullData.filter(regex=r'^occurreddate_', axis=1)`.

```python
categ_dict = {}
for i in dataCateg:
    if i not in categ_dict:
        categ_dict[i] = fullData.filter(regex=r'^'+i, axis=1)
```

```python

count = 0 
for i in categ_dict:
    print("type [" + i + "] has " + str(len(categ_dict[i].columns)) + " columns")

    count += len(categ_dict[i].columns)
    
print(count)
```
