## Split Dataset to three part: Training, evaluation and testing data

```python
import random
clientList = list(set(init_record.index.get_level_values(0)))
```

```python
random.shuffle(clientList)
```

```python
length = len(clientList)
ratio = int(length*0.1)
trainIDs = clientList[ratio:]
testIDs = clientList[:ratio]
train0 = init_record.drop(testIDs, level=0)
test = init_record.drop(trainIDs, level=0)
```

```python
print(train0.shape)
print(test.shape)
```

```python
ratio = int(len(trainIDs)*0.25)
trainRatio = trainIDs[ratio:]
evalRatio  = trainIDs[:ratio]
train = train0.drop(evalRatio, level=0)
evl = train0.drop(trainRatio,level=0)
```

```python
print(train.shape)
print(evl.shape)
print(test.shape)
```

## Construct feature and label

```python
import numpy as np
```

#### <span style="color:red">Notice! This function will change dataframe inplace, if running second time have to rerun the code of init_record  </span>

```python

def construct_feature_label(data):
    y = []
    clientList = list(set(data.index.get_level_values(0)))
    data['response_variable'] = data.groupby(level=0)['response_variable'].shift(-1)  
    data = data.groupby(level=0, as_index=False).apply(lambda group: group.iloc[:-1])
    last_month_y = list(data.groupby(level=0).apply(lambda group: group.iloc[-1])['response_variable'].astype('int32'))
    #print(data.loc["0000D172-EA88-432F-8235-9FAA00D29072"]['response_variable'])
    y = data['response_variable'].astype('int32')
    x = data.drop('response_variable',axis='columns')
    return x, y, last_month_y

x_train, y_train, y_train_last = construct_feature_label(train)
x_eval, y_eval, y_eval_last = construct_feature_label(evl)
```

```python
print(len(y_train))
print(x_train.shape)
print(len(y_eval))
print(x_eval.shape)
```

## Use tsfresh to extract features

```python
from tsfresh import extract_features
```

```python
x_train = x_train.reset_index()
x_eval = x_eval.reset_index()
```

```python
train_pca['_key_client_id'] = x_train['_key_client_id']
train_pca['_key_occurreddate_month'] = x_train['_key_occurreddate_month']
eval_pca['_key_client_id'] = x_eval['_key_client_id']
eval_pca['_key_occurreddate_month'] = x_eval['_key_occurreddate_month']
```

```python
train_extracted = extract_features(train_pca, column_id="_key_client_id", column_sort="_key_occurreddate_month")
```

```python
eval_extracted = extract_features(eval_pca, column_id="_key_client_id", column_sort="_key_occurreddate_month")
```

```python
train_extracted.filter(regex = r"0__value_count", axis=1)
```

```python
eval_extracted
```

```python
len(y_train_last)
```

```python
train_extracted.to_csv("ts_train_extracted.csv")
eval_extracted.to_csv("ts_eval_extracted.csv")
```

```python
import lightgbm
from sklearn.metrics import f1_score
```

```python
import re
train_ex = train_extracted.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', 'N', x))
eval_ex = eval_extracted.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', 'N', x))
```

```python
train_data = lightgbm.Dataset(train_ex, label=y_train_last)
eval_data = lightgbm.Dataset(eval_ex, label=y_eval_last)
```

```python
parameters = {
    'application': 'binary',
    'objective': 'binary',
    'metric': 'auc',
    'is_unbalance': 'true',
    'boosting': 'gbdt',
    'num_leaves': 31,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 20,
    'learning_rate': 0.05,
    'verbose': 0
}
```

```python
model = lightgbm.train(parameters,
                       train_data,
                       valid_sets=eval_data,
                       num_boost_round=2000)
```

```python
#x_eval
```

```python
y_hat = np.where(model.predict(eval_ex) < 0.5, 0, 1)  
f1_score(y_eval_last,y_hat)
```
