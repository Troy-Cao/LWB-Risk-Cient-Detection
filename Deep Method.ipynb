```python
# summarize the total number of clients (year)
index_dict = {}
for ind in init_record.index:
    if ind[0] not in index_dict:
        index_dict[ind[0]] = []
        index_dict[ind[0]].append(ind[1])
    else:
        index_dict[ind[0]].append(ind[1])

print(len(index_dict))
```

```python
## Get the information of records for each client
## Accroding to the summarization, 7380 clients have 12-month record
record_num = []
for value in index_dict.values():
    record_num.append(len(value))

# print(record_num) 

from collections import Counter

Counter(record_num)
```

```python
## Store the full-record clinet record in initial_data
initial_data = {}
for inde in index_dict.keys():
    if len(index_dict[inde]) == 12:
        initial_data[inde] = init_record.loc[inde]
print(len(initial_data))
```

```python
## Summarize the max incident number of each client
max_info_dict = {}
for key in initial_data.keys():
    max_value = max(initial_data[key]['Total_incident_count'])
    max_info_dict[key] = max_value
```

```python
## Summarize the Initial_data into negative and positive sample set
negative_sample = []
positive_sample = []
for key in max_info_dict.keys():
    if max_info_dict[key] == 0:
        negative_sample.append(initial_data[key])
    else:
        positive_sample.append(initial_data[key])
print(len(negative_sample))
print(len(positive_sample))
```

```python
import random
import torch

seed_num = 7
random.seed(seed_num)
torch.manual_seed(seed_num)

all_sample = random.sample(positive_sample, 2500) + random.sample(negative_sample, 2500)
print(len(all_sample))
```

```python
## Unsplit feature and label;
all_feature = []
all_label = []

for sample in all_sample:
    all_label.append(sample['response_variable'])
    temp_sample = sample.drop(['response_variable'], axis=1)
    all_feature.append(temp_sample)
    
print(len(all_feature))
print(len(all_label))
```

```python
feature_sample = []
label_sample = []


## k is the window size
def split_df(feature_data, label_data, k):
    
    df_list = []
    label_list = []
    i = 0
    while i+k < feature_data.shape[0]:
        temp = feature_data[i: i+k, :]
        df_list.append(temp)
        label_list.append(label_data[i+k])
        i += 1
    return df_list, label_list

for i in range(len(all_feature)):
    feature_temp_list, label_temp_list = split_df(all_feature[i].values, all_label[i].values, 4)
    for feature in feature_temp_list:
        feature_sample.append(feature)
    for label in label_temp_list:
        label_sample.append(label)
print(len(feature_sample))
print(len(label_sample))
```

### Model Construction

```python
import torch.nn as nn
import torch.functional as F

class MV_time_CNN(nn.Module):
    def __init__(self, input_dim, output_dim, window_size):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels = input_dim, out_channels = output_dim, kernel_size = (4, window_size[0]))
        self.conv2 = nn.Conv2d(in_channels = input_dim, out_channels = output_dim, kernel_size = (4, window_size[1]))
        self.conv3 = nn.Conv2d(in_channels = input_dim, out_channels = output_dim, kernel_size = (4, window_size[2]))
        self.full_connect = nn.Linear(in_features = 423, out_features = 1)
        self.dropout = nn.Dropout(p = 0.5)
        
    def forward(self, data):
        data = torch.unsqueeze(data, 1)
        hidden_layer1 = self.conv1(data)
        hidden_layer2 = self.conv2(data)
        hidden_layer3 = self.conv3(data)
        hidden_layer = torch.cat([hidden_layer1, hidden_layer2, hidden_layer3], dim=3)
        hidden_layer = hidden_layer.squeeze(1)
        hidden_layer = hidden_layer.squeeze(1)
        output = self.full_connect(hidden_layer)
        return output
```

```python
#defining the training method
def train(model, iterator, optimizer, criterion):
    
    epoch_loss = 0
    epoch_acc = 0
    epoch_F1_score = 0

    model.train()
    
    for batch in iterator:
        
        input_data = batch[0].to(device)
        labels = batch[1].to(device)

        optimizer.zero_grad()
        
        predictions = model(input_data).squeeze(1)
        
        loss = criterion(predictions,labels)
        
        acc = binary_accuracy(predictions, labels)
        f_score = f_score_measure(predictions, labels)

        loss.backward()
        
        optimizer.step()
        
        epoch_loss += loss.item()
        epoch_acc += acc.item()
        epoch_F1_score += f_score.item()
        
    return epoch_loss / len(iterator), epoch_acc / len(iterator), epoch_F1_score / len(iterator)
```

```python
#defining the validation method
def evaluate(model, iterator, criterion):
    
    epoch_loss = 0
    epoch_acc = 0
    epoch_F1_score = 0
    
    model.eval()

    
    with torch.no_grad():
    
        for batch in iterator:
            input_data = batch[0].to(device)
            labels = batch[1].to(device)

            predictions = model(input_data).squeeze(1)
            
            loss = criterion(predictions, labels)
            
            acc = binary_accuracy(predictions, labels)
            f_score = f_score_measure(predictions, labels)

            epoch_loss += loss.item()
            epoch_acc += acc.item()
            epoch_F1_score += f_score.item()

    return epoch_loss / len(iterator), epoch_acc / len(iterator), epoch_F1_score / len(iterator)
```

```python
#defining the method to calculate epoch time
import time

def epoch_time(start_time, end_time):
    elapsed_time = end_time - start_time
    elapsed_mins = int(elapsed_time / 60)
    elapsed_secs = int(elapsed_time - (elapsed_mins * 60))
    return elapsed_mins, elapsed_secs
```
